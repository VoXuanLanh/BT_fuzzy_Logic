```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
Distance = ctrl.Antecedent(np.arange(0, 51, 0.1), 'Distance')
Traffic_condition = ctrl.Antecedent(np.arange(0, 101, 1), 'Traffic_condition')
Demand_level = ctrl.Antecedent(np.arange(0, 101, 1), 'Demand_level')
Weather_condition = ctrl.Antecedent(np.arange(0, 11, 1), 'Weather_condition')
Customer_rating = ctrl.Antecedent(np.arange(0, 5.1, 0.5), 'Customer_rating')
Ride_punctuality = ctrl.Antecedent(np.arange(0, 101, 1), 'Ride_punctuality')
Ride_price = ctrl.Consequent(np.arange(0, 101, 10), 'Ride_price')
Customer_reward_points = ctrl.Consequent(np.arange(0, 25.1, 0.1), 'Customer_reward_points')
Distance['short'] = fuzz.trimf(Distance.universe, [0, 0, 3.1])
Distance['medium'] = fuzz.trimf(Distance.universe, [1.9, 4, 8.1])
Distance['long'] = fuzz.trimf(Distance.universe, [5.9, 10, 21])
Distance['very_long'] = fuzz.trimf(Distance.universe, [14.9, 35, 50])
Traffic_condition['low'] = fuzz.trimf(Traffic_condition.universe, [0, 0, 31])
Traffic_condition['medium'] = fuzz.trimf(Traffic_condition.universe, [19, 40, 71])
Traffic_condition['high'] = fuzz.trimf(Traffic_condition.universe, [59, 80, 100])
Demand_level['low'] = fuzz.trimf(Demand_level.universe, [0, 0, 31])
Demand_level['medium'] = fuzz.trimf(Demand_level.universe, [19, 50, 71])
Demand_level['high'] = fuzz.trimf(Demand_level.universe, [59, 75, 100])
Weather_condition['good'] = fuzz.trimf(Weather_condition.universe, [0, 0, 4])
Weather_condition['moderate'] = fuzz.trimf(Weather_condition.universe, [4, 6, 7])
Weather_condition['bad'] = fuzz.trimf(Weather_condition.universe, [7, 8.5, 10])
Customer_rating['poor'] = fuzz.trimf(Customer_rating.universe, [0.5, 2, 2.6])
Customer_rating['average'] = fuzz.trimf(Customer_rating.universe, [1.5, 3, 4.5])
Customer_rating['good'] = fuzz.trimf(Customer_rating.universe, [3.4, 4, 5])
Ride_punctuality['late'] = fuzz.trimf(Ride_punctuality.universe, [0, 0, 51])
Ride_punctuality['on_time'] = fuzz.trimf(Ride_punctuality.universe, [39, 60, 81])
Ride_punctuality['early'] = fuzz.trimf(Ride_punctuality.universe, [69, 80, 100])
Ride_price['low'] = fuzz.trimf(Ride_price.universe, [0, 0, 30])
Ride_price['medium'] = fuzz.trimf(Ride_price.universe, [29, 35, 50])
Ride_price['high'] = fuzz.trimf(Ride_price.universe, [45, 60, 70])
Ride_price['very_high'] = fuzz.trimf(Ride_price.universe, [70, 85, 100])
Customer_reward_points['none'] = fuzz.trimf(Customer_reward_points.universe, [0, 0.5, 1])
Customer_reward_points['low'] = fuzz.trimf(Customer_reward_points.universe, [1, 3, 5])
Customer_reward_points['moderate'] = fuzz.trimf(Customer_reward_points.universe, [4, 8, 12.5])
Customer_reward_points['high'] = fuzz.trimf(Customer_reward_points.universe, [10, 17.5, 25.1])
```

```python
rule1 = ctrl.Rule(Distance['short'] & Traffic_condition['low'] & Demand_level['low'], Ride_price['low'])
rule2 = ctrl.Rule(Distance['short'] & Traffic_condition['medium'] & Demand_level['high'], Ride_price['medium'])
rule3 = ctrl.Rule(Distance['medium'] & Traffic_condition['high'] & Demand_level['high'], Ride_price['high'])
rule4 = ctrl.Rule(Distance['long'] & Traffic_condition['medium'] & Weather_condition['good'], Ride_price['medium'])
rule5 = ctrl.Rule(Distance['long'] & Traffic_condition['high'] & Weather_condition['bad'], Ride_price['very_high'])
rule6 = ctrl.Rule(Distance['very_long'] & Traffic_condition['high'] & Demand_level['high'], Ride_price['very_high'])
rule7 = ctrl.Rule(Distance['medium'] & Traffic_condition['low'] & Demand_level['low'], Ride_price['medium'])
rule8 = ctrl.Rule(Distance['short'] & Traffic_condition['high'] & Weather_condition['bad'], Ride_price['high'])
rule9 = ctrl.Rule(Distance['very_long'] & Weather_condition['bad'], Ride_price['very_high'])
rule10 = ctrl.Rule(Distance['medium'] & Traffic_condition['medium'] & Weather_condition['moderate'], Ride_price['medium'])
rule11 = ctrl.Rule(Customer_rating['good'] & Ride_punctuality['early'], Customer_reward_points['high'])
rule12 = ctrl.Rule(Customer_rating['average'] & Ride_punctuality['on_time'], Customer_reward_points['moderate'])
rule13 = ctrl.Rule(Distance['long'] & Traffic_condition['high'] & Ride_punctuality['on_time'], Customer_reward_points['high'])
rule14 = ctrl.Rule(Distance['medium'] & Traffic_condition['medium'] & Customer_rating['good'], Customer_reward_points['moderate'])
rule15 = ctrl.Rule(Customer_rating['poor'] & Ride_punctuality['late'], Customer_reward_points['none'])
rule16 = ctrl.Rule(Distance['very_long'] & Weather_condition['bad'] & Customer_rating['good'], Customer_reward_points['high'])
rule17 = ctrl.Rule(Distance['short'] & Customer_rating['average'] & Ride_punctuality['on_time'], Customer_reward_points['low'])
rule18 = ctrl.Rule(Distance['long'] & Traffic_condition['high'] & Ride_punctuality['late'], Customer_reward_points['low'])
rule19 = ctrl.Rule(Distance['medium'] & Weather_condition['moderate'] & Customer_rating['good'], Customer_reward_points['moderate'])
```

```python
from math import e
fuzzy_system = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18, rule19])
fuzzy_system_sim = ctrl.ControlSystemSimulation(fuzzy_system)
fuzzy_system_sim.input['Distance'] = 2
fuzzy_system_sim.input['Traffic_condition'] = 70
fuzzy_system_sim.input['Demand_level'] = 3
fuzzy_system_sim.input['Weather_condition'] = 8
fuzzy_system_sim.input['Customer_rating'] = 4
fuzzy_system_sim.input['Ride_punctuality'] = 80
fuzzy_system_sim.compute()
Ride_price.view(sim=fuzzy_system_sim)
Customer_reward_points.view(sim=fuzzy_system_sim)

print("giá đi xe: ", fuzzy_system_sim.output['Ride_price'])
def get_Ride_price(price):
  if price < 30:
    return 'low'
  elif price < 50 and price > 29:
    return 'medium'
  elif price < 70 and price > 45:
    return 'high'
  else:
    return 'very_high'
print(f"giá đi xe: {get_Ride_price(fuzzy_system_sim.output['Ride_price'])}")

print(" Điểm thưởng: ", fuzzy_system_sim.output['Customer_reward_points'])
def get_Customer_reward_points(points):
  if points < 1:
    return 'none'
  elif points < 5 and points > 1:
    return 'low'
  elif points <= 12 and points > 4:
    return 'moderate'
  else:
    return 'high'
print(f"điểm thưởng: {get_Customer_reward_points(fuzzy_system_sim.output['Customer_reward_points'])}")
```
